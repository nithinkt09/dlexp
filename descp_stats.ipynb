```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import pandas as pd
```

```python
# accuracy: relationship b/w measurement and actual values
# precision: certainty of each measurement (inversely related to variance)
# resolution: number of data points per unit measurement
```

```python
N = 1000

x = np.linspace(-60,60,N)
gausdist = stats.norm.pdf(x)

plt.plot(x,gausdist)
print(sum(gausdist) * np.diff(x[:2]))
plt.show()
```
