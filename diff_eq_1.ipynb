```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
A = np.array([[0.5, 0.5, 0.25], [0.25, 0, 0.25], [.25,.5,.5]] )
x_today = np.array([[1],[0],[0]])

x_tom = A @ x_today

wether_data = np.zeros((3,50))
for k in range(50):
  xtomorrow = A @x_today
  wether_data[:,k] = xtomorrow.reshape(3)
  x_today = xtomorrow

wether_data
```

```python
plt.plot(wether_data[0,:])
plt.plot(wether_data[1,:])
plt.plot(wether_data[2,:])
plt.grid()
plt.legend(['Sunny', 'Cloudy', 'Rainy'])
plt.xlabel('Time')
plt.ylabel('Probability')
plt.show()
```
