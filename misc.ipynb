```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
```

```python
# use gpu if available
device = "cude" if torch.cuda.is_available() else "cpu"

print(device)
```

```python
X = torch.arange(0, 1, 0.02).unsqueeze(dim=1)

weight = 0.8
bias = 0.3

y = weight * X + bias

train_len = int(0.8 * len(X))

X_train, y_train = X[:train_len], y[:train_len]
X_test, y_test = X[train_len:], y[train_len:]

```

```python
def visualize_data(X_train=X_train, y_train=y_train, X_test=X_test, y_test=y_test, predictions=None):
    plt.figure(figsize=(10,7))

    plt.scatter(X_train, y_train, s=8, c='g', label='training')
    plt.scatter(X_test, y_test, s=8, c='b', label='testing')

    if predictions is not None:
        plt.scatter(X_test, predictions, c='r', label='predictios', s=3)
```

```python
class LinearRegressionModel(nn.Module):
    def __init__(self):
        super(LinearRegressionModel, self).__init__()

        # using nn.Linear() for creating model parameters
        self.linear_layer = nn.Linear(in_features=1, out_features=1)

    def forward(self, x):
        return self.linear_layer(x)

torch.manual_seed(32)
```

```python
model_linear_regression = LinearRegressionModel()

loss_function = nn.L1Loss()
optimizer =  optim.SGD(model_linear_regression.parameters(), lr=0.01)
num_epochs = 1000

epoch_count = []
loss_values = []
test_loss_values = []


for epoch in range(num_epochs):
    model_linear_regression.train()
    optimizer.zero_grad()
    model_predictions = model_linear_regression(X_train)
    loss = loss_function(model_predictions, y_train)
    loss.backward()
    optimizer.step()

    model_linear_regression.eval()
    with torch.inference_mode():
        test_predictions = model_linear_regression(X_test)
        test_loss = loss_function(test_predictions, y_test)
        if epoch % 100 == 0:
            epoch_count.append(epoch)
            loss_values.append(loss)
            test_loss_values.append(test_loss_values)

```

```python
with torch.inference_mode():
    y_preds = model_linear_regression(X_test)

visualize_data(predictions=y_preds)
```
