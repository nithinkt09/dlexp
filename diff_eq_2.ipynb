```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from scipy.special import factorial
```

```python
x = np.linspace(-10,10,20000)
y = np.sin(x)

plt.plot(x,y,'k', linewidth=2)
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.grid(True)


# first order taylor expansion
p = [1,0]
yT1 = np.polyval(p,x)
plt.plot(x,yT1,'b--', linewidth=1.2, label='1st order')

# third order taylor epansion
p = [-1/factorial(3), 0, 1, 0]
yt3 = np.polyval(p,x)
plt.plot(x,yt3,'r--', linewidth=1.2, label='3rd order')

# fifth order taylor expansion
p = [1/factorial(5), 0, -1/(factorial(3)), 0, 1, 0]
yt5 = np.polyval(p,x)
plt.plot(x,yt5,'g--',linewidth=1.2, label='5th order')

# seventh order taylor expansion
p = [-1/factorial(7), 0, 1/factorial(5), 0, -1/factorial(3), 0, 1, 0]
yt7 = np.polyval(p,x)
plt.plot(x,yt7, 'c--', linewidth=1.2, label='7th order')

plt.legend()

plt.show()
```
