```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# function
def fx(x):
    return 5 * x**3 + 2 * x - 4

def deriv(x):
    return 15 * x ** 2  + 2
```

```python
# plot the function and its derivative

# range of values for x or independent variable
x = np.linspace(-5, 5, 2001)


# plot the function and it's derivative
plt.plot(x, fx(x), x, deriv(x))
plt.grid()
plt.xlim(x[[-5, 5]])
plt.xlabel('x')
plt.ylabel('f(x)')
```

```python
# simple learning algorithm
local_min = np.random.choice(x, 1) # use one of the inputs of x for initial local min

# learning parameters
learning_rate = 0.01 # steps that the algorithm must take in each iteration
training_epochs = 100 # number of iterations the algorithm must take

for i in range(training_epochs):
    grad = deriv(local_min)
    local_min = local_min = learning_rate * grad

local_min
```

```python
# plot the results
plt.plot(x, fx(x), x, deriv(x))
plt.plot(local_min, deriv(local_min),'ro')
plt.plot(local_min, fx(local_min), 'ro')

plt.grid()
plt.xlim(x[[-5, 5]])
plt.xlabel('x')
plt.ylabel('f(x)')
```
