MNIST dataset comprises of 70,000 greyscale images of handwritten digits from 0 to 9 with each image sized at 28x28 pixels and each image has respective labeles indicating the digit. pixel values range from 0 to 255.

image as a set of pixels, one way to view image is by characterizing it by intensity and spatial coordinates, I = {p_xy, (x,y) | 0<= x, y <= 255}
where p_xy is the pixel intensity at point (x,y)
by emphasizing and considering pixels as unordered collections of pixels it can further the abilties of set theory and using them to focus on intrinsic properties of image independent of the arrangement of the pixels
in addition, by considering layers as objects and transformations (pooling or convolutions) as morphisms, it would be easier to analyze the compositional structure of the network which helps in abstraction and also how different components may interact


```python
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import datasets, transforms
```

```python

class IMGSetRepresentation(Dataset):
    def __init__(self, train=True):
        self.data = datasets.MNIST(
            root=".",
            train=train,
            download=True,
            transform=transforms.ToTensor()
        )

    def __len__(self):
        return len(self.data)

    def __get__item(self, idx):
        image, label = self.data[idx] #retrieve the image at the specified location
        image = image.squeeze(0) # removes the first dimension of the tensor if it is of size 1, which is number of channels generally 1 in this dataset since it is greyscale
        pixel_set = {
            ((x,y), image[x, y].item())
            for x in range(28)
            for y in range(28)
        }

        return pixel_set, label


train_dataset = IMGSetRepresentation(train=True)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
```
