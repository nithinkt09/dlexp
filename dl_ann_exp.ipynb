```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
import numpy as np
from IPython import display
display.set_matplotlib_formats('svg')
```

deep learning model :
- guess a solution
- compute the error or mistakes or far off the solution or predictions are from true values
- use the errors and use it as learning parameter to modify the parameter, here it will be done by findining the minimum error to miniize the distance between predictions and actual values


perceptron: unit in neural networks that takes inputs, calculates summed weights of inputs, applies an activation function and gives an output

weighted_sum or averaging_machine = dot_product(input_data * weights) #linear operation: addition and scalar multiplication

weighted_sum is passed into a sig function or an activation function (a nonlinear fuction)

output = activation_function(linear_weighted_sum + bias)


feature_space: a geometric representation of input data, where each feature could be considered as a point on respective axis and each observation or output of the respective input can  be considered as a coordinate

seperating_hyperplane: a boundary that binarizes or categorizes data which can then be used as a decision boundary or a notion of prediction

categories of model outputs
- discrete/categorical/binary: outputs are mapped to a selective number of classess or groups or values
- numeric/continuous: outputs are mapped into R_n(real space)



forward propagation: input data transformed into outputs or predictions
 -    y = activation_function(input*weights + bias)

error: difference or how far off the prediction is compared to actual value

in general, continuous erros are used for training and classification for evaluation

erros are then used to  generate loss which can be done for both continuous or discrete values. where derivatives of loss are used for gradients and adjusting parameters

cost is average of loss for all the input

backpropagation: adjusts the weights and bias with respect to loss

goal of deep learning model can be summarized as finding the set of weights that minimizes the all the losses or the average cost
