```python
import sqlite3
from datetime import datetime
```

```python

# Connect to the SQLite database (or create it if it doesn't exist)
conn = sqlite3.connect('/content/notes.db')
cursor = conn.cursor()

# Create a table for storing notes if it doesn't exist
cursor.execute('''CREATE TABLE IF NOT EXISTS notes (
                    id INTEGER PRIMARY KEY AUTOINCREMENT,
                    date TEXT,
                    note_name TEXT,
                    keywords_subjects TEXT,
                    keywords_concepts TEXT,
                    text TEXT
                )''')

conn.commit()

# Function to add a new note
def add_note(note_name, keywords_subjects, keywords_concepts, text):
    date = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    cursor.execute("INSERT INTO notes (date, note_name, keywords_subjects, keywords_concepts, text) VALUES (?, ?, ?, ?, ?)",
                   (date, note_name, keywords_subjects, keywords_concepts, text))
    conn.commit()
    print(f"Note '{note_name}' added to the database.")
```

```python

# Example usage
add_note("Matrix Inverse Rule", "Matrix, Function", "Inverse, Bijection, Linearly Dependent", "In a matrix with linearly dependent columns...")
```
